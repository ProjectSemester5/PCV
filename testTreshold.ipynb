```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('straw3.jpg')  # Ganti dengan jalur gambar Anda
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Terapkan Gaussian Blur untuk mengurangi noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Deteksi tepi dengan ambang yang berbeda
edges1 = cv2.Canny(blurred, 50, 100)  # Threshold rendah
edges2 = cv2.Canny(blurred, 100, 200)  # Threshold sedang
edges3 = cv2.Canny(blurred, 150, 300)  # Threshold tinggi

# Buat subplot untuk menampilkan gambar
plt.figure(figsize=(10, 8))

# Gambar asli
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Konversi ke RGB untuk matplotlib
plt.title("Original Image")
plt.axis("off")

# Hasil deteksi tepi dengan ambang rendah
plt.subplot(2, 2, 2)
plt.imshow(edges1, cmap='gray')
plt.title("Edges with Low Threshold")
plt.axis("off")

# Hasil deteksi tepi dengan ambang sedang
plt.subplot(2, 2, 3)
plt.imshow(edges2, cmap='gray')
plt.title("Edges with Medium Threshold")
plt.axis("off")

# Hasil deteksi tepi dengan ambang tinggi
plt.subplot(2, 2, 4)
plt.imshow(edges3, cmap='gray')
plt.title("Edges with High Threshold")
plt.axis("off")

# Tampilkan semua subplot
plt.tight_layout()
plt.show()
```
