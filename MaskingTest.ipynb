```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Fungsi untuk masking merah dan hijau dalam LAB


# Fungsi untuk masking merah saja dalam LAB
def mask_red_lab(image):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

    lower_red = np.array([0, 150, 0])
    upper_red = np.array([255, 255, 255])
    mask_red = cv2.inRange(lab, lower_red, upper_red)
    return mask_red

# Fungsi untuk masking merah saja dalam RGB
def mask_red_rgb(image):
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    lower_red1 = np.array([200, 0, 0])
    upper_red1 = np.array([255, 80, 80])
    mask_red = cv2.inRange(rgb, lower_red1, upper_red1)

    return mask_red

# Baca gambar
image = cv2.imread('straw3.jpg')  # Ganti dengan jalur gambar Anda

# Masking pada setiap gambar# Gambar 1: Masking merah dan hijau (LAB)
mask2 = mask_red_lab(image)        # Gambar 2: Masking merah saja (LAB)
mask3 = mask_red_rgb(image)        # Gambar 3: Masking merah saja (RGB)
# Tampilkan hasil dalam subplot
plt.figure(figsize=(12, 8))

# Gambar asli
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')


# Hasil Masking 2
plt.subplot(1, 3, 2)
plt.imshow(mask2, cmap='gray')
plt.title('Mask Red Only (LAB)')
plt.axis('off')

# Hasil Masking 3
plt.subplot(1, 3, 3)
plt.imshow(mask3, cmap='gray')
plt.title('Mask Red Only (RGB)')
plt.axis('off')

# Tampilkan plot
plt.tight_layout()
plt.show()

```
